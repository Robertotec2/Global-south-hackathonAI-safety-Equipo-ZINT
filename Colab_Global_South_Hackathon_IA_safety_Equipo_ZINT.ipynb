```python
import os
import whisper
import pandas as pd
from pathlib import Path
import time

def encontrar_carpeta():
    candidatos = [p for p in Path('.').rglob('audios_yucatan') if p.is_dir()]
    if candidatos:
        return candidatos[0]
    return None

def main():
    AUDIO_DIR = encontrar_carpeta()
    OUTPUT_CSV = "resultados_colab.csv"

    if AUDIO_DIR is None:
        print("❌ ¡Ayuda! No encuentro la carpeta 'audios_yucatan'.")
        print("Esto es lo que veo en el directorio actual:", os.listdir('.'))
        return

    print(f"✅ ¡Encontrada! La carpeta está en: {AUDIO_DIR.resolve()}")
    print("Cargando modelo Whisper en la GPU T4... 🚀")
    model = whisper.load_model("medium")

    # Definimos qué archivos buscar
    archivos = [f for f in AUDIO_DIR.iterdir() if f.suffix.lower() in ['.mp3', '.wav', '.m4a', '.mpeg', '.mpg']]

    if not archivos:
        print("⚠️ No encontré archivos de audio (.mp3, .wav, .m4a, .mpeg, .mpg) en la carpeta.")
        return

    resultados = []

    # 3. Procesar cada audio y transcribirlo
    for archivo in archivos:
        print(f"⏳ Escuchando: {archivo.name} ...")
        start = time.time()

        try:
            resultado = model.transcribe(str(archivo), language="es")
            texto = resultado["text"].strip()
            estado = "Éxito"

            tiempo = time.time() - start
            print(f"✅ Listo ({tiempo:.1f}s): {texto[:60]}...")

        except Exception as e:
            texto = str(e)
            estado = "Error"
            print(f"❌ Falló {archivo.name}: {e}")

        resultados.append({
            "Nombre_Archivo": archivo.name,
            "Transcripcion_Whisper": texto,
            "Estado": estado
        })

    # 4. Exportar resultados al CSV
    df = pd.DataFrame(resultados)
    df.to_csv(OUTPUT_CSV, index=False, encoding="utf-8-sig")
    print(f"\n🎉 ¡Proceso terminado! Resultados guardados en: {OUTPUT_CSV}")

if __name__ == "__main__":
    main()
```
